```python
#Créer un objet indiquant avec des boleens si le taux de chomage pour
#le 1er trimesttre 2020 est supérieur a 7.1

data_t1_2020 = data['2020-T1']>7.1
data_t1_2020
```

```python
#Créer un objet indiquant avec des booléens si le taux de chômage pour le 1ere trimestre
#2019 est supérieur à 7.1

data_t1_2019 = data['2019-T1']>7.1
data_t1_2019
```

```python
#Créer un objet indiquant avec des booléens si le taux de chômage pour le 1ere trimestre
#2019 est supérieur à 7.1

data_t1_2018 = data['2018-T1']>7.1
data_t1_2018
```

```python
#Combiner les trois objets Series avec l’operateur « & » 
#et assigné le résultat à la variable

data_7_1 = (data_t1_2020 & data_t1_2019 & data_t1_2018)
data_7_1
```

```python
#data_7_1 est une liste. Assigné là à un dataframe dont le 
#nom de colonne sera « booleen »

df = pd.DataFrame({'booléen' : data_7_1})
df
```

```python
#Charger le fichier « pole_emploi.xlsx » après l’avoir ouvert et 
#visualiser et l’assigner à la variable « p_emploi ».

nom_col = ['REG', 'LIB', 'PE2018']
p_emploi = pd.read_excel('//home//simplon//Downloads//pole_emploi.xlsx', names=nom_col, sheet_name='Data', skiprows=3)
p_emploi
```

```python
#Joindre les data et p_emploi au seins d’un seul et unique dataframe en faisant
#correspondre les régions par leur codes. Utiliser pd.merge() pour réaliser cette jointure.

data_merge=pd.merge(p_emploi, data, on='REG', how='right')
data_merge
```

```python
#Créer une colonne quantile dans p_emploi qui prendra les valeurs [‘q25’, ‘q50’, ‘q75’,
#q100’] en fonction du quartile auquel appartient le nombre de pôle emploi.


data_merge['quantile']=pd.qcut(data_merge['PE2018'], 4, labels=['q25','q50','q75','q100'])
data_merge
```

```python
data_pivot=pd.pivot_table(data_merge, values='2020-T1', index=['LIBREG'], columns=['quantile'])
data_pivot
```
