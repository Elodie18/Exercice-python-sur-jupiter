```python
#import de la librairi pandas avec l'allias pd

import pandas as pd

# Création d'un dataframe

df = pd.DataFrame({'colonne1' : [1.1, 2.7, 5.3], 'colonne2' : [2, 10,9], 'colonne3' : ['oui','non','non]}']},
                      index = ['a', 'b', 'c'])
print (df)
```

```python
#Import de données depuis un fichié Excel dans un dataframe

nom_colonnes = ['consommation','cylindres', 'déplacement', 'puissance', 'poids', 'accélération', 'année', 'origine', 'nom']

data = pd.read_excel('Downloads//autompg.xlsx',
                     index_col=None, header=None, names=nom_colonnes, skiprows=1)

#Afficher les premières lignes du dataframe
print (data.head())
```

```python
#Afficher les dimensions du dataframe

print(data.shape)
print(data.shape[0])
print(data.shape[1])
```

```python
# Déscription des données
print(data.describe())
```

```python
#Informations sur les données
print(data.info())
```

```python
#Fréquences croisés
print(pd.crosstab(data['cylindres'],data['année']))
```

```python
#pourcentage croisé
print(pd.crosstab(data['cylindres'],data['année'], normalize='index'))
```

```python
#Accés à une ou plusieurs colonne
print(data[['nom','consommation']].head())
```

```python
#Afficher les valeurs 10 à 12
print(data['nom'][10:13])
```

```python
#Accées à la valeur de la première ligne, première colonne
print(data.iloc[0, 0])
#Accées à la valeur de la dernière ligne, première colonne
print(data.iloc[-1, 0])
```

```python
#Toutes les colonnes avec les lignes de 0 à 4
print(data.iloc[0:5,:])
```

```python
#Liste des voitures dont qui sont de l'année 80
print(data.loc[data['année']==80,:].head())
```

```python
#Moyenne de consommation des voitures de 8 cylindres
(data.loc[data['cylindres']==8, "consommation"].mean())
```

```python
# Véhicule année 70 ayant la plus forte puissances 
data70=data.loc[data['année']==70]
max70=data.loc[data['année']==70,'puissance'].max()
print(data70.loc[data['puissance']==max70,'nom'])
```

```python
#Pour que les graphiques apparaissent dans le notebook
%matplotlib inline

#Importation de la librairie
import matplotlib.pyplot as plt

#Histogramme de la puissance
data.hist(column='puissance')
```
