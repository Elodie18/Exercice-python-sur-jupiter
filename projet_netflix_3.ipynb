```python
donneescat= donnees[donnees['listed_in'].notna()][['show_id','listed_in']]
show_listed_in = pd.DataFrame(columns = ['show_id','listed_in'])
for i in donneescat.index:
    liste1 = donneescat["listed_in"][i].split(", ")
    liste2 = [donneescat["show_id"][i]] * len(liste1) 
    df = pd.DataFrame({'show_id': liste2, 'listed_in': liste1}, columns = ['show_id', 'listed_in'])
    show_listed_in = show_listed_in.append(df, ignore_index = True)
show_listed_in
```

```python
listed_in = show_listed_in["listed_in"].drop_duplicates()
listed_in = listed_in.reset_index(drop=True)
listed_in = listed_in.reset_index()
listed_in = listed_in.rename(columns={"index":"listed_in_id","listed_in":"listed_in"})
show_listed_in = show_listed_in.merge(listed_in).drop("listed_in", axis = 1)
show_listed_in
```

```python
listed_in
```

```python
donneesdir = donnees[donnees['director'].notna()][['show_id','director']]
show_director = pd.DataFrame(columns = ['show_id','director'])
for i in donneesdir.index:
    liste1 = donneesdir["director"][i].split(", ")
    liste2 = [donneesdir["show_id"][i]] * len(liste1) 
    df = pd.DataFrame({'show_id': liste2, 'director': liste1}, columns = ['show_id', 'director'])
    show_director = show_director.append(df)

director_id = show_director["director"].drop_duplicates()
director_id = director_id.reset_index(drop=True)
director_id = director_id.reset_index()
director_id = director_id.rename(columns={"index":"director_id","director":"director"})
show_director = show_director.merge(director_id).drop("director", axis = 1)
show_director  
```

```python
director_id
```

```python
donneescast = donnees[donnees['cast'].notna()][['show_id','cast']]
show_cast = pd.DataFrame(columns = ['show_id','cast'])
for i in donneescast.index:
    liste1 = donneescast["cast"][i].split(", ")
    liste2 = [donneescast["show_id"][i]] * len(liste1) 
    df = pd.DataFrame({'show_id': liste2, 'cast': liste1}, columns = ['show_id', 'cast'])
    show_cast = show_cast.append(df)
show_cast  
cast_id = show_cast["cast"].drop_duplicates()
cast_id = cast_id.reset_index(drop=True)
cast_id = cast_id.reset_index()
cast_id = cast_id.rename(columns={"index":"cast_id","cast":"cast"})
show_cast = show_cast.merge(cast_id).drop("cast", axis = 1)
show_cast
```

```python
cast_id
```

```python
donneescountry = donnees[donnees['country'].notna()][['show_id','country']]
show_country = pd.DataFrame(columns = ['show_id','country'])
for i in donneescountry.index:
    liste1 = donneescountry["country"][i].split(", ")
    liste2 = [donneescountry["show_id"][i]] * len(liste1) 
    df = pd.DataFrame({'show_id': liste2, 'country': liste1}, columns = ['show_id', 'country'])
    show_country = show_country.append(df)
show_country  
country_id = show_country["country"].drop_duplicates()
country_id = country_id.reset_index(drop=True)
country_id = country_id.reset_index()
country_id = country_id.rename(columns={"index":"country_id","country":"country"})
show_country = show_country.merge(country_id).drop("country", axis = 1)
show_country
```

```python
country_id
```

```python
show_duration = pd.DataFrame(columns = ['show_id','duration'])

for i in donnees.index:
    liste1 = donnees["duration"][i].split(", ")
    liste2 = [donnees["show_id"][i]] * len(liste1) 
    df = pd.DataFrame({'show_id': liste2, 'duration': liste1}, columns = ['show_id', 'duration'])
    show_duration = show_duration.append(df)
show_duration = show_duration["duration"].drop_duplicates()


show_duration = show_duration.reset_index(drop=True)

show_duration = show_duration.reset_index()

show_duration = show_duration.rename(columns={"index":"duration_id","duration":"duration"})


show_duration.duration = show_duration.duration.str.replace("Seasons","")
show_duration.duration = show_duration.duration.str.replace("Season", "")
show_duration.duration = show_duration.duration.str.replace("min", "")
show_duration.duration = show_duration.duration.astype("int")

show_duration
```

```python
catalogue = donnees.drop(columns=['listed_in','country','director','cast'])
catalogue.date_added =pd.to_datetime(catalogue['date_added'])
catalogue.duration = catalogue.duration.str.replace("Seasons","")
catalogue.duration = catalogue.duration.str.replace("Season", "")
catalogue.duration = catalogue.duration.str.replace("min", "")
catalogue.duration = catalogue.duration.astype("int")
```
